# Analysis of damages in CO2-electrolysers by post-mortem X-ray scans
## General description

- Miniature CO2-electrolysers were built in the same way as for the TOMCAT-beamtime
- A X-ray CT was taken before operation at the nanotom
- 4 samples were operated at differrent current densities (25, 50, 100, 200 mA/cm2), but with the same charge, i.e. by reducing operation time
- A CT is taken after half the operation time and after the end of operation
- cells are operated in the lab, operation is interrupted and cell disconnected to acquire imaging, i.e. in-situ, but not operando

## Qualitative analysis
### A_1
test case @3.5V constant potential
#### preop
- membrane and CL intact
#### postop 1
- delamination from fibers, but no larger damage in membrane
- CL damage not obvious -> used Cu filter for later experimenets
#### postop 2
- multiple "bubbly" delaminations
- clear CL breakages
